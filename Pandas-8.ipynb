Leo solo las columnas que necesito, solo porque ya s√© como es el .csv

```python
df_contents = df_contents.loc[ df_contents.title.notna() ]
```

Paso la revision_timestamp a DateTime para poder aplicarle la mediana despu√©s.

```python
df_contents["revision_timestamp"] = pd.to_datetime( df_contents["revision_timestamp"] )
```

Agrupo seg√∫n el primer car√°cter del t√≠tulo.

```python
agrupados_por_primer_caracter = df_contents.groupby( ( df_contents["title"].str[0] ) )
```

Le aplico la mediana al "revision_timestamp" de cada conjunto groupby.

```python
agrupados_por_primer_caracter.agg( { "revision_timestamp": ["median"] } )
```
