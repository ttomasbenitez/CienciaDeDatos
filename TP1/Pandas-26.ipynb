```python
df_logs = pd.read_csv("/content/drive/MyDrive/datosTP1/logs.csv")
```

Veo como es el dataframe.

```python
df_logs.head()
```

Veo que tipos de action hay, encontrando 'block' y 'reblock'.


```python
df_logs.action.unique()
```

Creo un nuevo dataframe con las filas cuya action sea block o reblock.

```python
opciones = ["block", "reblock"]
bloqueos = df_logs.loc[ df_logs["action"].isin(opciones) ]
```

```python
bloqueos.head()
```

```python
bloqueos_validos = bloqueos.loc[ bloqueos.title.notna() ]
```

Agrupo los usuarios bloqueados repetidos.

```python
bloqueados_agrupados = bloqueos_validos.groupby(["title"])
```

Los paso a un dataframe a los bloqueados_agrupados y cuento cuantas veces fue bloqueado cada usuario.


```python
usuarios_bloqueados = bloqueados_agrupados.agg( { "item_id": ["count"] } )
```

```python
usuarios_bloqueados.columns = ["veces_bloqueado"]
```

```python
mas_bloqueado = usuarios_bloqueados.veces_bloqueado.max()
```

```python
usuarios_bloqueados[ usuarios_bloqueados.veces_bloqueado >= mas_bloqueado ]
```
